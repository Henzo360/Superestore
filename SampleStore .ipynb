# Handling Duplicates

```python
duplicates_mask = sample.duplicated()
duplicates_count = duplicates_mask.sum()
print('Number of duplicates:', duplicates_count)
```

```python
duplicates = sample.duplicated(keep=False)

print("Duplicated rows:")
sort_df = sample[duplicates].sort_values("Postal Code")
sort_df
```

sample_df = sample.drop_duplicates()
